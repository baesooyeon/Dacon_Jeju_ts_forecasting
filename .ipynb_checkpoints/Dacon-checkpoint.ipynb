```python
# 컬럼명 변경
df.rename(columns = {"supply(kg)" : "supply", "price(원/kg)" : "price"}, inplace = True)
```

### 농산물 별 EDA

```python
f, axes  = plt.subplots(2,3)
f.set_size_inches((20,15))

# 날짜별로 정렬해주고 그래프 그리기!
df = df.sort_values(by = "timestamp")


axes[0,0].plot(df[df["item"] == "TG"]["timestamp"], df[df["item"] == "TG"]["price"])
axes[0,0].set_title("감귤 가격 그래프")

axes[0,1].plot(df[df["item"] == "CR"]["timestamp"], df[df["item"] == "CR"]["price"])
axes[0,1].set_title("당근 가격 그래프")

axes[0,2].plot(df[df["item"] == "CB"]["timestamp"], df[df["item"] == "CB"]["price"])
axes[0,2].set_title("양배추 가격 그래프")

axes[1,0].plot(df[df["item"] == "RD"]["timestamp"], df[df["item"] == "RD"]["price"])
axes[1,0].set_title("무 가격 그래프")

axes[1,1].plot(df[df["item"] == "BC"]["timestamp"], df[df["item"] == "BC"]["price"])
axes[1,1].set_title("브로콜리 가격 그래프")
```

품목별로 모델을 생성하면 좋을 거 같다!

```python
index_ = []
for i in range(3):
    for j in range(2):
        index_.append([i,j])
        
f, axes = plt.subplots(3,2, figsize = (13,7))

# 세로 간격을 조정 : hspace
# 가로간격을 조정 : wspace
plt.subplots_adjust(hspace = 0.5)

# 자동으로 세로간격 조정

for i in range(5):
    x = index_[i][0]
    y = index_[i][1]
    
    food = df["food"].unique()[i]
    data = df[df["food"] == df["food"].unique()[i]]
    
    axes[x,y].plot(data["timestamp"], data["supply"])
    
    axes[x,y].set_title(f"{food} 수요량 그래프")
```

```python
index_ = []
for i in range(3):
    for j in range(2):
        index_.append([i,j])
        
f, axes = plt.subplots(3,2, figsize = (13,13))

# 세로 간격을 조정 : hspace
# 가로간격을 조정 : wspace
plt.subplots_adjust(hspace = 0.5)

# 자동으로 세로간격 조정

for i in range(5):
    x = index_[i][0]
    y = index_[i][1]
    
    food = df["food"].unique()[i]
    data = df[df["food"] == df["food"].unique()[i]]
    
    
    axes[x,y].plot(data["timestamp"], data["supply"])
    axes[x,y].plot(data["timestamp"], data["price"])
    axes[x,y].set_title(f"{food} 수요량 및 가격 그래프")
```

```python
# 스케일링해서 다시 비교

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df_ss = ss.fit_transform(df[["supply", "price"]])

df["supply_ss"] = np.transpose(df_ss)[0]
df["food_ss"] = np.transpose(df_ss)[1]


index_ = []
for i in range(3):
    for j in range(2):
        index_.append([i,j])
        
f, axes = plt.subplots(3,2, figsize = (13,13))

# 세로 간격을 조정 : hspace
# 가로간격을 조정 : wspace
plt.subplots_adjust(hspace = 0.5)

# 자동으로 세로간격 조정

for i in range(5):
    x = index_[i][0]
    y = index_[i][1]
    
    food = df["food"].unique()[i]
    data = df[df["food"] == df["food"].unique()[i]]
    
    
    axes[x,y].plot(data["timestamp"], data["supply_ss"])
    axes[x,y].plot(data["timestamp"], data["food_ss"])
    axes[x,y].set_title(f"{food} 수요량 및 가격 그래프")
```

```python
f, axes = plt.subplots(2,3,figsize =(13,7))

for idx, col in zip([[0,0],[0,1],[0,2],[1,0],[1,1]],df.food.unique()):
    idx_x = idx[0]
    idx_y = idx[1]
    sns.boxplot(data = df[df["food"] == col], x = "food", y = "price", ax = axes[idx_x, idx_y])
```

양배추, 당근, 무는 이상치값들이 많은것을 알 수 있다. 감귤의 최댓값(이상치)가 발생한 날짜를 확인하고 어떤 이슈가 있었는지 봐도 좋을 거같다.
